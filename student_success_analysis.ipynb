```python
import sqlite3
```

```python
conn = sqlite3.connect(
    "projects/student_success_dashboard/student_success.db"
)
```

```python
clean_df.to_sql(
    "student_metrics",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT *
FROM student_metrics
LIMIT 5
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT ROUND(AVG(Graduation_Rate),2)
AS Avg_Graduation_Rate
FROM student_metrics
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT ROUND(AVG(Retention_Rate),2)
AS Avg_Retention_Rate
FROM student_metrics
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Enrollment)
AS Total_Enrollment
FROM student_metrics
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    State,
    ROUND(SUM(Enrollment),0) AS Total_Enrollment
FROM student_metrics
GROUP BY State
ORDER BY Total_Enrollment DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Institution,
    Graduation_Rate
FROM student_metrics
WHERE Graduation_Rate IS NOT NULL
ORDER BY Graduation_Rate DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Institution,
    Median_Earnings
FROM student_metrics
WHERE Median_Earnings IS NOT NULL
ORDER BY Median_Earnings DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    State,
    ROUND(AVG(Graduation_Rate),2) AS Avg_Graduation_Rate,
    ROUND(AVG(Retention_Rate),2) AS Avg_Retention_Rate,
    ROUND(AVG(Median_Earnings),0) AS Avg_Earnings
FROM student_metrics
GROUP BY State
ORDER BY Avg_Graduation_Rate DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    Institution,
    Enrollment,
    Graduation_Rate,
    Retention_Rate,
    Median_Earnings
FROM student_metrics
WHERE Institution LIKE '%Maryland%'
   OR Institution LIKE '%Towson%'
   OR Institution LIKE '%Morgan%'
   OR Institution LIKE '%Baltimore County%'
"""

pd.read_sql(query, conn)
```
